## 4. Cohort Analysis

### Creating Cohorts based on Tenure

### Let's begin by treating each unique tenure length, 1 month, 2 month, 3 month...N months as its own cohort.

```python
no_churn = df.groupby(['Churn','tenure']).count().transpose()['No']
yes_churn = df.groupby(['Churn','tenure']).count().transpose()['Yes']
```

```python
churn_rate = 100 * yes_churn / (no_churn+yes_churn)
```

```python
churn_rate.transpose()['customerID']
```

Now we have Churn Rate per tenure group 1-72 months.

```python
plt.figure(figsize=(10,4),dpi=200)
churn_rate.iloc[0].plot()
plt.title('Churn Rate per tenure group')
plt.ylabel('Churn Percentage');
```

### Broader Cohort Groups

Creating new column called Tenure Cohort that creates 4 separate categories:

- 0-12 Months
- 24-48 Months
- 12-24 Months
- Over 48 Months

```python
def cohort(tenure):
    if tenure < 13:
        return '0-12 Months'
    elif tenure < 25:
        return '12-24 Months'
    elif tenure < 49:
        return '24-48 Months'
    else:
        return "Over 48 Months"
```

```python
df['Tenure Cohort'] = df['tenure'].apply(cohort)
```

```python
df.head(10)[['tenure','Tenure Cohort']]
```

Scatterplot of Total Charges versus Monthly Charts,colored by Tenure Cohort.

```python
plt.figure(figsize=(10,4),dpi=200)
sns.scatterplot(data=df,x='MonthlyCharges',y='TotalCharges',hue='Tenure Cohort', linewidth=0.5,alpha=0.5,palette='Dark2')
plt.title('Total Charges versus Monthly Charts');
```

Count plot showing the churn count per cohort.

```python
plt.figure(figsize=(10,4),dpi=200)
plt.title('Churn Count per cohort')
sns.countplot(data=df,x='Tenure Cohort',hue='Churn');
```

Count Plots showing counts per Tenure Cohort, separated out by contract type and colored by the Churn hue.

```python
plt.figure(figsize=(10,4),dpi=200)
sns.catplot(data=df,x='Tenure Cohort',hue='Churn',col='Contract',kind='count')
plt.title('Count Plots showing counts per Tenure Cohort, separated out by contract type and colored by the Churn hue.');
```

## 5. Predictive Modeling

Let's explore 3 different tree based methods: 
- Single Decision Tree
- Random Forest 
- AdaBoost 

### Data preparation

Separating data into X features, y label and creating dummy variables,

```python
X = df.drop(['Churn','customerID'],axis=1)
X = pd.get_dummies(X,drop_first=True)
```

```python
y = df['Churn']
```

Performing train test split, holding out 10% of the data for testing.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

### Single Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(max_depth=6)
```

```python
dt.fit(X_train,y_train)
```

```python
preds = dt.predict(X_test)
```

### Single Tree Model Performance

```python
from sklearn.metrics import accuracy_score,plot_confusion_matrix,classification_report
```

```python
print(classification_report(y_test,preds))
```

```python
import warnings
warnings.filterwarnings("ignore")
plot_confusion_matrix(dt,X_test,y_test);
```

```python
imp_feats = pd.DataFrame(data=dt.feature_importances_,index=X.columns,columns=['Feature Importance']).sort_values("Feature Importance")
```

```python
plt.figure(figsize=(14,6),dpi=200)
sns.barplot(data=imp_feats.sort_values('Feature Importance'),x=imp_feats.sort_values('Feature Importance').index,y='Feature Importance')
plt.xticks(rotation=90)
plt.title("Feature Importance for Decision Tree");
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(12,8),dpi=150)
plot_tree(dt,filled=True,feature_names=X.columns);
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=100)
```

```python
rf.fit(X_train,y_train)
```

```python
preds = rf.predict(X_test)
```

### Random Forest Performance

```python
print(classification_report(y_test,preds))
```

```python
import warnings
warnings.filterwarnings("ignore")
plot_confusion_matrix(dt,X_test,y_test);
```

### Ada Boosted Trees

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ada_model = AdaBoostClassifier()
```

```python
ada_model.fit(X_train,y_train)
```

```python
preds = ada_model.predict(X_test)
```

### Ada Boosted Trees Performance
